# Iterating over arrays

````python
a = np.arange(6).reshape(2,3)
a
````

````python
for x in a:
    print(x, end=' ')
````

````python
a = np.arange(6).reshape(2,3)
for x in np.nditer(a):                       #default order = 'K'  keep the existing order
    print(x, end= ' ')
````

````python
a = np.arange(8).reshape(4,2)
a
````

````python
for x in np.nditer(a):               #default order = 'K'  keep the existing order
    print(x, end = ' ')
````

````python
for x in np.nditer(a, order='K'):    #order = 'K'  keep the existing order
    print(x, end = ' ')
````

````python
for x in np.nditer(a, order='C'):    #order = 'C'  force C-order iteration
    print(x, end = ' ')
````

````python
a = np.arange(8).reshape(4,2)        #order = 'F'  force Fortran-order iteration
for x in np.nditer(a, order='F'):
    print(x, end = ' ')
````

````python
a = np.arange(8).reshape(4,2)
for x in np.nditer(a):
    print(x, end= ' ')
````

````python
a
````

````python
a.T
````

````python
a = np.arange(8).reshape(4,2)
for x in np.nditer(a.T):
    print(x, end= ' ')
````

````python
a = np.arange(8).reshape(4,2)
for x in np.nditer(a.T, order='C'):
    print(x, end= ' ')
````

## a.T is a view, not a copy

## underlying memory order stays the same

## np.nditer iterates in memory-efficient order, not logical index order

````python
a = np.arange(8).reshape(4,2)
for x in np.nditer(a.T, order='F'):
    print(x, end= ' ')
print()
for x in np.nditer(a.T, order='C'):
    print(x, end= ' ')
````

# View  vs  Copy

````python
array2 = np.arange(2*5).reshape(2,5)
array2
````

````python
subset = array2[:,2:]   # slicing returns a view not a copy
subset
````

````python
subset[1,2] = 100   # modifying the view will also modify the original array
subset
````

````python
array2
````

````python
array2 = np.arange(2*5).reshape(2,5)
print(array2)
print()
subset = array2[:, 2:].copy()  # Now slicing actually creates a copy
subset[1,2] = 100
print(subset) 
print()
print(array2)
````

# Conclusion

## NumPy Arrays

A **NumPy array** is an **N-dimensional array designed for homogeneous data**.  
This means every element in the array must have the **same data type**.

Example:

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print(arr)
```

---

## NumPy Data Types (dtypes)

NumPy arrays store elements with a specific **data type (`dtype`)**.

Common types include:

- `int`
- `float`
- `object` (used for strings or mixed types)

Example:

```python
arr = np.array([1, 2, 3])
print(arr.dtype)
```

---

## Homogeneous Data

NumPy arrays hold **homogeneous data** (same data type).

You can check the data type using:

```python
arr.dtype
```

Example:

```python
arr = np.array([1, 2, 3])
print(arr.dtype)   # int64 (may vary by system)
```

---

## Vectorization (Element-wise Operations)

**Vectorization** allows NumPy to perform operations **element-wise across arrays** without writing explicit loops.

Example:

```python
arr = np.array([1, 2, 3])

result = arr * 2
print(result)
```

Output:

```
[2 4 6]
```

---

## Broadcasting

**Broadcasting** allows NumPy to perform operations on arrays of **different shapes** by automatically expanding the smaller array so the shapes are compatible.

Example:

```python
arr = np.array([1, 2, 3])

result = arr + 5
print(result)
```

Output:

```
[6 7 8]
```

NumPy automatically broadcasts the scalar `5` across the array.

---

## Matrix Multiplication (Dot Product)

Matrix multiplication in NumPy can be performed using the **`@` operator**.

Example:

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

C = A @ B
print(C)
```

---

## Matrix Transposition

Transposing a matrix swaps its rows and columns.

Use the `.T` attribute:

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print(A.T)
```

---

## Scalars vs 1D vs 2D Arrays

NumPy indexing can return **different types of results** depending on how you slice.

Example:

```python
arr = np.array([[1, 2],
                [3, 4]])
```

| Expression | Result |
|-------------|--------|
| `arr[1,1]` | Scalar |
| `arr[1, :]` | 1D array |
| `arr[:, 1]` | 1D array |

---

## Scalar

A **scalar** refers to a single basic Python value, such as:

- `int`
- `float`
- `string`

Example:

```python
value = arr[1,1]
print(type(value))
```

---

## 1D Arrays

A **1D array** has only **one axis**.

Example:

```python
arr1 = np.array([1, 2, 3, 4])
```

Operations along axes:

```python
arr1.sum(axis=0)   # valid
```

Using `axis=1` will produce an **out-of-bounds error** because a 1D array only has one axis.

---

## Key Takeaways

- NumPy arrays store **homogeneous data**
- The data type is accessed using **`array.dtype`**
- **Vectorization** enables fast element-wise operations
- **Broadcasting** allows operations on arrays with different shapes
- Matrix multiplication uses **`@`**
- Transpose uses **`.T`**
- Indexing determines whether you get a **scalar, 1D array, or 2D array**
