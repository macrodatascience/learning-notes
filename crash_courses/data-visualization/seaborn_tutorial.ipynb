### Basic Setup

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Optional set a theme for better aesthetics
sns.set_theme(style="whitegrid")
````

### Loading Data (we can use one of the built-in datasets)

````python
# Load the example dataset
tips = sns.load_dataset("tips")
print(tips.head())
````

### 1️⃣ Relational Plots

### Scatter Plot

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time") # , size="size"
plt.show()
````

### Line Plot

````python
sns.lineplot(data=tips, x="size", y="tip", errorbar=None)
plt.show()
````

### 2️⃣ Distribution Plots

### Histogram & KDE

````python
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.show()
````

### KDE Plot

````python
sns.kdeplot(data=tips, x="total_bill", hue="sex", fill=True)
plt.show()
````

### ECDF Plot

````python
sns.ecdfplot(data=tips, x="total_bill", hue="sex")
plt.show()
````

### 3️⃣Categorical Plots

### Bar Plot

````python
sns.barplot(data=tips, x="day", y="total_bill", estimator=sum)
plt.show()
````

### Count Plot

````python
sns.countplot(data=tips, x="day", hue="sex")
plt.show()
````

### Box Plot

````python
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
plt.show()
````

### Violin Plot

````python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.show()
````

### Strip Plot

````python
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, jitter=True)
plt.show()
````

### 4️⃣ Matrix Plots

````python
flights = sns.load_dataset("flights")
flights
````

````python
pivot_table = flights.pivot(index="month", columns="year", values="passengers")
pivot_table
````

### Heatmap

````python
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.show()
````

### Cluster Map

````python
sns.clustermap(pivot_table, cmap="coolwarm", standard_scale=1)
plt.show()
````

### 5️⃣ Multi-variable relationships - Multi-Plot Grids 

    (Pair Plots, Joint Plots, Facet Grids)

### Pair Plots

````python
sns.pairplot(tips, hue="sex")
plt.show()
````

### Joint Plot

````python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex", color="g")
plt.show()
````

### Facet Grid

````python
g = sns.FacetGrid(tips, col="sex", row="time", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
plt.show()
````

### 6️⃣ Styling & Customization 

````python
sns.set_theme(style="darkgrid", palette="muted")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day", size="size", sizes=(20, 200))
plt.title("Tips vs Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
````
