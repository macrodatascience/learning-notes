# Git version-control system allows distributed projects (multiple people can work on the same project simultaneously, each version controlled project is called a repository

    Git repos allow work to be centralized in a single folder

    Git keeps track of the changes to any of the files in the repo on a constant basis

## Git works best with files (text) and not with other objects like word documents, excel files etc. 

## Git vs Github  (Local vs remote)

## Maintain a .gitignore file

    To avoid tracking unnecessary folders like data, __pycache__, .ipynb_checkpoints etc. 

## Installing Git

```
pip install git 
git --version
```

## Configuring Git

``` 
git config --global user.name "your name"
git config --global user.email "youremail@email.com"
```

## Creating a new Git repository

### from scratch 

```
git init
git remote add origin https://github.com/username/repository-name.git
```

## Cloning an existing repository

```
git clone https://github.com/username/repository-name.git
```

## Tracking changes to files

```

git add filename  # to stage changes 
git status
git rm --cached <filename> # to unstage changes

```

## Commits

```
git commit -m "Add filename with initial commit"
git add filename
git commit -m "made changes to filename"
```

## Syncing local and remote repositories 

```
git fetch origin main  # shows the list of changes to the remote repository but doenst update the files in the local repo

git pull origin main   # incorporates all the changes to the remote repository to the local repo

git push origin main   # pushes changes specifically to main on the remote repository
```

## In prior years, the primary branch of a repo was called master. 

## In recent versions of Git, the primary branch has been switched to main. 

## Feature branches  (use descriptive names to the feature branch)

```
git checkout -b my-feature-branch-updating-credit-metrics

git add filename
git commit -m "commit message for the credit metrics changes"
git status # to see if all the changes have been staged, and a summary of what has been committed 
```

## Merges

```
git checkout main
git merge my-feature-branch-updating-credit-metrics
```

## Conflicts

```
git fetch  # allows the fetch all the feature branches
git diff alice-feature-branch # shows the difference between your branch and alice-feature-branch
git merge alice-feature-branch 
git add sometext.text
git commit -m "resolved merge conflict by combining Alice's and my changes"
git checkout main 
git merge -my-feature-branch-updating-credit-metrics
git push
```

## Best Practices 
```
1. Commit often
2. Write clear commit messages
3. Use feature branches to segregate your work and this keeps main branch stable and bug-free for others using it before any new changes are tested and vetted
4. Name branches in a standardized way
5. Keep your branches in sync with your main branch
6. Push and pull between local and remote repositories frequently
```

## Commands for Troubleshooting
```
1. git status  # identify unstaged changes, staged changes, and untracked files
2. git diff    # view differences between commits, branches, or the working directory and index
3. git log     # review the commit history to trace when changes were made
4. git blame   # determine who last modified each line of a file and in which commit
5. git show    # inspect changes in a specific commit
6. git branch  # list, create, or delete branches to manage different lines of deployment
```

## How git works under the hood 

1. DAGs

    Every commit in a Git repo is a point, or node, in a DAG that points to its parent commit. These interconnected nodes create a treelike structure that facilitates the tracing of the lineage of each piece of code. Since the commits are directionally connected, it's far easier for Git to search for specific points in it's history than it would be if the commits were stored separately or with directionless connections.

2. Hashing

    Git uses hash table to quickly reference and locate commit data without searching through every single one.
