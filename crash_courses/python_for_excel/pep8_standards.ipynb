# Function-Definitions

Function naming conventions:

- lowercase
- words separated by underscores
- include a descriptive docstring

```python
def convert_to_celsius(degrees, source="fahrenheit"):
    """
    Convert temperature to Celsius.

    Parameters
    ----------
    degrees : float
        Temperature value to convert.
    source : str
        Temperature scale ("fahrenheit" or "kelvin").

    Returns
    -------
    float
        Temperature converted to Celsius.
    """

    if source.lower() == "fahrenheit":
        return (degrees - 32) * 5.0 / 9.0

    if source.lower() == "kelvin":
        return degrees - 273.15

    raise ValueError("Source must be 'fahrenheit' or 'kelvin'")
```

# Naming-Conventions

Variables should use **lowercase names with underscores**.

```python
celsius = convert_to_celsius(100, source="fahrenheit")
celsius
```

# Spacing-and-Comments

Guidelines:

- Inline comments should have **two spaces before the `#`**
- Use spaces around **mathematical operators**
- Do **not** add spaces inside parentheses or brackets
