````python
import datetime as dt
from pytz import timezone
````

````python
my_date = dt.date(2025, 12, 25)
my_time = dt.time(20, 40)
my_datetime = dt.datetime(2025, 12, 25, 20, 40)
my_timedelta = dt.timedelta(days=4)

curr = dt.datetime.now()
prev = curr - dt.timedelta(days=4)

print(my_date)
print(my_time)
print(my_datetime)
print(my_timedelta)

print(curr)
print(prev)
print(curr-prev)
````

# Converting datetime Objects to Strings

```
    %Y  'yyyy'
    %m  'mm'
    %d  'dd'  Zero-padded
    %H  'HH'  Zero-padded 24 hour clock
    %M  'MM'  Zero-padded
    %S  'SS'  Zero-padded
    %B  'full month mmmm' 
    %b  'abbrev month mmm'
    %A  'full week day name dddd' 
    %a  'abbrev week day name ddd'
```

````python
date_str = my_date.strftime('%Y-%m-%d')
print(date_str)
````

````python
datetime_str = my_datetime.strftime('%Y-%m-%d %H:%M')
print(datetime_str)
````

# Extracting datetime Objects from Strings

````python
datetime_str = '2025-12-20 14:40'

my_new_datetime = dt.datetime.strptime(datetime_str, '%Y-%m-%d %H:%S')
my_new_datetime
````

````python
datetime_str2 = my_new_datetime.strftime('%Y-%m-%d')
datetime_str2
````

# Working with timezones

````python
# pip install pytz
````

````python
naive_datetime = dt.datetime(2025, 12, 28, 12,10)
NY_datetime = timezone('America/New_York').localize(naive_datetime)
NY_datetime
````

````python
from pytz import all_timezones
print(all_timezones[:10])
````

````python
timezone_unaware_datetime = dt.datetime(2025, 12, 24, 12, 30, 30)
ny_timezone = timezone('America/New_York')

timezone_aware_datetime = ny_timezone.localize(timezone_unaware_datetime)
print(timezone_aware_datetime)

london_timezone = timezone('Europe/London')
london_timezone_datetime = london_timezone.localize(timezone_unaware_datetime)
print(london_timezone_datetime)
````
