```python
import pandas as pd

def normalize_fx(df, base_col="USD", pair_col="pair", rate_col="rate"):
    """
    Normalize FX rates so that a given currency is the base currency
    Default is the USD base currency.

    Parameters
    ----------
    df : pandas.DataFrame
    pair_col : column with currency pair (ex: EURUSD)
    rate_col : column with rate
    base_col : currency to use as base (default: USD)

    Returns
    -------
    pandas.DataFrame with columns:
        pair, rate (with the given base currency)
    """

    df = df.copy()

    # split currencies
    df["base"] = df[pair_col].str[:3]
    df["quote"] = df[pair_col].str[3:]

    new_pairs = []
    new_rates = []

    for base, quote, rate in zip(df["base"], df["quote"], df[rate_col]):

        if base == base_col:
            new_pairs.append(base + quote)
            new_rates.append(rate)

        elif quote == base_col:
            new_pairs.append(quote + base)
            new_rates.append(1 / rate)

        else:
            # if base_col not present, keep original
            new_pairs.append(base + quote)
            new_rates.append(rate)

    result = pd.DataFrame({
        "pair": new_pairs,
        "rate": new_rates
    })

    return result
```

```python
exchange_rates = {
    "EURUSD": 1.1,
    "USDJPY": 110.0,
    "GBPUSD": 1.3,
    "USDKRW": 1200.0,
    "USDCHF": 0.9,
    "USDCAD": 1.25,
    "AUDUSD": 0.75,
    "NZDUSD": 0.7
}

df = pd.DataFrame(exchange_rates.items(), columns=["pair", "rate"])

normalized_df = normalize_fx(df, base_col='USD')

print(normalized_df)
```
