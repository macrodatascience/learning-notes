## Seaborn Library Tutorial

#### Seaborn is a Python library built on top of Matplotlib, designed for creating attractive and informative statistical graphics. It simplifies complex visualizations and provides a high-level API for data visualization. Below is a comprehensive guide to get started with Seaborn.

### Ensure dependencies like numpy, pandas, matplotlib, and scipy are installed

````python
import seaborn as sns
import matplotlib.pyplot as plt
````

### Basic Plotting

````python
# Load dataset
data = sns.load_dataset("iris")

# Create a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

### Customization with Matplotlib

````python
# Seaborn integrates seamlessly with Matplotlib, allowing further customization

# Add a title using Matplotlib
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title("Iris Sepal Dimensions")
plt.show()
````

### Styling and Themes

#### Seaborn offers several themes to enhance plot aesthetics. 

Use the **set_style()** method to apply themes like **darkgrid, whitegrid, dark, white, or ticks**

````python
sns.set_style("whitegrid")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

````python
sns.set_style("darkgrid")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

````python
sns.set_style("dark")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

````python
sns.set_style("white")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

### To temporarily change the style, use **axes_style() within a with statement**:

````python
with sns.axes_style("whitegrid"):
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
````

### Color Palettes

````python
# Seaborn provides tools to customize color palettes for better visualization

# Use a diverging color palette
palette = sns.color_palette("PiYG", 11)
sns.palplot(palette)
plt.show()
````

````python
# To set a default palette for all plots

sns.set_palette("vlag")
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

### Advanced Plots

### Seaborn supports a variety of plot types for different use cases

**Scatter Plot**: Visualize relationships between two variables.

**Box Plot**: Show distributions and outliers.

**Heatmap**: Visualize matrix data with color coding.

````python
sns.scatterplot(x="sepal_length", y="sepal_width", data=data)
plt.show()
````

````python
sns.boxplot(x="species", y="sepal_width", data=data)
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only numeric columns
correlation = data.select_dtypes(include='number').corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5,
    square=True
)

plt.title("Correlation Heatmap", fontsize=14)
plt.show()
````

### Multi-Plot Grids

### Seaborn allows creating grids of plots using **FacetGrid** or **PairGrid**

````python
# FacetGrid example
g = sns.FacetGrid(data, col="species")
g.map(sns.scatterplot, "sepal_length", "sepal_width")
plt.show()
````

### Regression and Statistical Plots

### Seaborn simplifies regression analysis with **regplot()** and **lmplot()**

````python
tips = sns.load_dataset("tips")
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
````
